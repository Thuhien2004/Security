## 1. Tên gọi

- Còn gọi là transposition cipher hay permutation cipher.

- Ý tưởng: không thay ký tự, chỉ đổi chỗ (hoán vị) các ký tự trong plaintext theo một quy tắc (khóa). Do đó tần suất ký tự được giữ nguyên, chỉ vị trí thay đổi.

## 2. Thuật toán mã hóa, giải mã

Giả sử khóa là một hoán vị 
𝑃 trên 𝑛 vị trí (ví dụ P=[3,1,4,2] nghĩa là vị trí 1 → 3, 2 → 1, 3 → 4, 4 → 2).

Mã hóa theo Block:

- Chia plaintext thành các block độ dài 𝑛.  Nếu block cuối thiếu, điền ký tự padding (ví dụ X).

- Với mỗi block 
𝐵=[𝑏1,𝑏2,...,𝑏𝑛] , tạo ciphertext block 𝐶 sao cho C[i]=B[P[i]] (tùy cách định nghĩa; quan trọng là dùng cùng quy ước khi giải mã).

(Hoặc ngược: C[P[i]]=B[i] — chỉ cần nhất quán.)

Pseudocode (định nghĩa: P là mảng chỉ số 1..n):


```python
encrypt_block(B[1..n], P[1..n]):
  for i from 1..n:
    C[i] = B[P[i]]
  return C
```

- GIẢI MÃ

Cần hoán vị nghịch đảo 𝑃−1 sao cho [P[i]]=i.

Giải mã mỗi block: B[i] = C[P^{-1}[i]] (hoặc theo công thức nghịch đảo tương ứng).

Tìm nghịch đảo P⁻¹ (ví dụ P=[3,1,4,2]):

P[1]=3 ⇒ P⁻¹[3]=1

P[2]=1 ⇒ P⁻¹[1]=2

P[3]=4 ⇒ P⁻¹[4]=3

P[4]=2 ⇒ P⁻¹[2]=4
=> P⁻¹ = [2,4,1,3]

## 3. kHÔNG gian khóa

Nếu khóa là hoán vị trên 𝑛 vị trí thì số khóa = n! (n giai thừa).

Ví dụ: n=5 → 5! = 120; n=10 → 10! ≈ 3.6 triệu; n=20 → ~2.43×10^18.

Tuy nhiên độ an toàn thực tế phụ thuộc vào cách dùng (cỡ block, padding, kết hợp với substitution,...).

## 4. Cách phá mã

Lưu ý: vì hoán vị giữ tần suất ký tự, nên phân tích tần suất đơn ký tự không hiệu quả — cần dùng cấu trúc vị trí, bigram/trigram, cấu trúc từ, và kỹ thuật tìm kiếm.

A. Các kịch bản tấn công

- Ciphertext-only (chỉ có ciphertext): khó nhất, nhưng không bất khả thi với cỡ block nhỏ hoặc dữ liệu dài.

- Known-plaintext (có một vài cặp plaintext-ciphertext): dễ phá — suy ra hoán vị.

- Chosen-plaintext: cực kỳ dễ — chỉ gửi một plaintext với mẫu/đánh dấu để đọc trực tiếp hoán vị.

B. Chiến lược tấn công ciphertext-only (thực tế)

1.Xác định kích thước block (n)

- Thử các n khả dĩ (2..k), cho mỗi n:

- Chia ciphertext thành cột (hoặc block) tương ứng và tính Index of Coincidence (IoC) hoặc đo sự "gần giống" với IoC của tiếng (ví dụ tiếng Anh ~0.066). Nếu ghép đúng cột, IoC trong từng cột tiến về IoC ngôn ngữ.

- Hoặc tính tổng tần suất bigram/ngram khi sắp xếp theo giả định n — giá trị cao hơn gợi ý n đúng.

2.Đặt vấn đề thành bài toán sắp xếp cột

- Với columnar transposition, ciphertext tương ứng các cột bị hoán vị. Ta cần tìm thứ tự cột sao cho khi đọc từng hàng sẽ tạo ra nhiều từ hợp lệ.

3.Sử dụng thuật toán tìm kiếm heuristics

- Hill-climbing / simulated annealing / genetic algorithms để tìm hoán vị tối ưu.

- Định nghĩa hàm điểm (score): dùng mô hình ngôn ngữ (log-likelihood), tần suất bigram/trigram hoặc điểm từ vựng (số lượng từ hợp lệ tiếng) để đánh giá một hoán vị là “tốt”.

- Khởi tạo hoán vị ngẫu nhiên → lặp: hoán đổi cột, nếu score tăng giữ lại; nếu không, có thể chấp nhận đôi khi (simulated annealing).

- Đây là phương pháp rất hiệu quả cho transposition cipher khi ciphertext đủ dài.

4.Dùng crib / từ đoán (crib dragging)

- Nếu nghi ngờ plaintext có từ cố định (ví dụ “THE”, “ATTACK”), thử đặt crib vào mọi vị trí hợp lý trong block, suy ra các phép hoán vị tương thích, kiểm thử.

- Kết hợp với phân tích n-gram

- Sau mỗi hoán vị thử, tính điểm n-gram (trigram/bigram) trên toàn văn; chọn hoán vị có điểm cao nhất.

C. Phá thủ công đơn giản (small n)

- Nếu n rất nhỏ (n ≤ 6..7), brute-force thử mọi hoán vị (n! thử) có thể khả thi. Với mỗi hoán vị, giải mã và kiểm tra xem văn bản có ý nghĩa (từ trong từ điển) hay không.
